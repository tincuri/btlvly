```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
theta = np.linspace(5,8,30)
r = 2* theta / (2*np.pi)
np.append(0,r)
theta2 = np.linspace(5,8,7)
r2 = 2* theta2 / (2*np.pi)
r2
```

```python
for i in range(len(r2)-1):
    subtheta = np.linspace(theta2[i],theta2[i+1], 10)
    subr = 2*subtheta / (2*np.pi)
    r3 = np.append(np.array([0,r2[i]]),subr)
    theta3= np.append(np.array([0, theta2[i], theta2[i+1]]),subtheta)
    print(r3)
```

```python

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(theta, r)
for i in range(len(r2)-1):
    subtheta = np.linspace(theta2[i],theta2[i+1], 10)
    subr = r2[i] * subtheta / subtheta
    r3 = np.append(np.array([0,r2[i]]),subr)
    theta3= np.append(np.array([0, theta2[i]]),subtheta)
    ax.plot(theta3,r3, color='black')
#ax.set_rmax(2.9)
ax.set_rticks([0.5, 1, 1.5, 2])  # Less radial ticks
ax.set_xticklabels([])  # Move radial labels away from plotted lineõ
ax.set_yticklabels([])
ax.grid(True)
plt.show()
```
