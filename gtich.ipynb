```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
theta = np.arange(6,9,0.1)
r = theta / (2*np.pi)
r
```

```python

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(theta, r)
ax.set_rmax(1.2)
ax.set_rticks([0.5, 1, 1.5, 2])  # Less radial ticks
ax.set_xticklabels([])  # Move radial labels away from plotted lineõ
ax.set_yticklabels([])
ax.grid(True)
theta=np.arrange(7,9,0.1)

plt.show()
```
